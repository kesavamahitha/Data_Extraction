```python
import re
```

```python
List = ['Year(s)', 'Ref.']
df = dfsrc.copy()
pattern = "(\[[0-9]*[a-z]*\])"

df["Peak"] = df["Peak"].apply(lambda x: re.sub(pattern, "", str(x)))
df["All Time Peak"] = df["All Time Peak"].apply(lambda x: re.sub(pattern, "", str(x)))
df["Tour title"] = df["Tour title"].apply(lambda x: re.sub(pattern, "", str(x)))

df["Year_Start"] = df["Year(s)"].apply(lambda x : x[0:4])
df["Year_End"] = df["Year(s)"].apply(lambda x : x[-4:])

df.drop(List, axis =1, inplace=True)
df.head()
```

```python
for cols in df.columns:
    if df[cols].dtype == "object":
        df[cols] = df[cols].apply(lambda x: x.replace("$","").replace("nan","0"))
        df[cols] = df[cols].apply(lambda x: re.sub(pattern, "", str(x)))\

df["Peak"].astype(int)
df["All Time Peak"].astype(int)

for cols in df.columns[3],df.columns[4],df.columns[7],df.columns[8],df.columns[9],df.columns[10]:
    df[cols] = df[cols].apply(lambda x: str(x).replace(",","").replace(",",""))
    df[cols] = df[cols].astype(int)
df.head()
```
